# Relationship between price and other variables

```python
plt.figure(figsize=(12, 8))
sns.pairplot(house_df[['price', 'area', 'bedrooms', 'bathrooms']])
plt.suptitle('Pairplot of Price, Area, Bedrooms, and Bathrooms', y=1.02)
plt.show()
```

# Boxplot of price by furnishing status

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=house_df, x='furnishingstatus', y='price', palette='Set2')
plt.title('Price Distribution by Furnishing Status')
plt.xlabel('Furnishing Status')
plt.ylabel('Price')
plt.show()
```
